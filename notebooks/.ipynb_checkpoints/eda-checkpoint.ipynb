```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/data.csv')
```

```python
data['Datetime'] = data['Date'].astype('datetime64[ns]') + data['Hour'].astype('timedelta64[h]')
```

```python
data.head()
```

```python
data.tail()
```

```python
data.plot(x='Datetime', y='Price ($/MWh)')
plt.show()
```

```python
data.plot(x='Datetime', y='Regional Power Demand (MW)')
plt.show()
```

```python
data.plot.scatter(x='Hour', y='Price ($/MWh)')
plt.show()
```
